```python
X = housing_data.drop(columns=['Price'])  # Features
y = housing_data['Price']  # Target
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('scaler', StandardScaler()),
])
```

```python
housing_pp = my_pipeline.fit_transform(X)
```

Checking predictions using linear Regressor

```python
from sklearn.linear_model import LinearRegression
linearmodel = LinearRegression()
linearmodel.fit(housing_pp,y)
```

```python

data = X.iloc[:5] 
label = y.iloc[:5]
```

```python
prepdata = my_pipeline.transform(data)
```

```python
linearmodel.predict(prepdata)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
mse = linearmodel.predict(housing_pp)
rmse = np.sqrt(mean_squared_error(y, mse))
```

```python
rmse
```

```python

from sklearn.model_selection import cross_val_score
scores = cross_val_score(linearmodel, housing_pp , y , scoring="neg_mean_squared_error", cv=10)
rmse_scores = np.sqrt(-scores)
```

```python
rmse_scores
```

```python
def print_scores(scores):
    print("Scores:", scores)
    print("Mean: ", scores.mean())
    print("Standard deviation: ", scores.std())
```

```python
print(print_scores(rmse_scores))
```

Checking predicitons using decision tree regressor

```python
from sklearn.tree import DecisionTreeRegressor
DecisionTreeModel = DecisionTreeRegressor()
DecisionTreeModel.fit(housing_pp,y)
```

```python
DecisionTreeModel.predict(prepdata)
```

```python
scores2 = cross_val_score(DecisionTreeModel, housing_pp , y , scoring="neg_mean_squared_error", cv=10)
rmse_scores2 = np.sqrt(-scores2)
```

```python
rmse_scores2
```

```python
print(print_scores(rmse_scores2))
```

Checking predictions using random Forest

```python
from sklearn.ensemble import RandomForestRegressor
RandomForestModel = RandomForestRegressor()
RandomForestModel.fit(housing_pp,y)
```

```python
RandomForestModel.predict(prepdata)
```

```python
scores3 = cross_val_score(RandomForestModel, housing_pp , y , scoring="neg_mean_squared_error", cv=10)
rmse_scores3 = np.sqrt(-scores3)
```

```python
rmse_scores3
```

```python
print(print_scores(rmse_scores3))
```
